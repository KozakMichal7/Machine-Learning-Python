# Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn
```

# Wygenerowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
print(df_raw)

# Utworzenie kopii danych
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
# LabelEncoder
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()  # Używamy tej klasy do kodowania zmiennej objaśnianej
le.fit(df['bought'])
le.transform(df["bought"])
```

```python
le.fit_transform(df['bought'])
```

```python
le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought'])
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
```

```python
print(encoder.transform(df[['size']])) # Info o pozycji w macierzy, na której stoja jedynki
# Miejsce jedynki wskazuje rozmiar: 'L', 'M', 'XL'
```

```python
df['size']
```

```python
# Wyrzucimy pierwszą kolumnę, żeby obserwacje nie były zależne (Skoro rozmiary to tylko M, L, XL, to jak wiemy, że to nie M ani XL, to z atomatu wiemy, że to L - nie potrzeba tego zaznaczać w macierzy)
encoder = OneHotEncoder(drop='first', sparse=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

# Standaryzacja - **StandardScaler**

```python
# std() - pandas nieobciążony
# std() - numpy obciążony

print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.5f}")

(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
